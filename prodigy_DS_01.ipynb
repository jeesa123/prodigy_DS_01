```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path= 'C:/Users/jeesa/Downloads/API_SP.POP.TOTL_DS2_en_csv_v2_6298256.csv'
df = pd.read_csv(file_path, skiprows=4)
```

```python
df
```

```python
# Select only the relevant columns: Country Name and population in 2022
df_cleaned = df[["Country Name", "2022"]].dropna()
```

```python
# Remove any rows with 0 or negative population
df_cleaned = df_cleaned[df_cleaned["2022"] > 0]
```

```python
# Get the top 20 countries/regions by population in 2022
top_20 = df_cleaned.sort_values(by="2022", ascending=False).head(20)
top_20
```

```python

plt.figure(figsize=(12, 8))
plt.barh(top_20["Country Name"], top_20["2022"], color='skyblue')
plt.xlabel("Population in 2022")
plt.title("Top 20 Most Populous Countries/Regions in 2022")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df_cleaned["2022"], bins=30, color='orange', edgecolor='black')
plt.xlabel("Population in 2022")
plt.ylabel("Number of Countries/Regions")
plt.title("Histogram of Population Distribution (2022)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
